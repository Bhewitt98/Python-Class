```python
#Imports the app and Book class from the Bookstore file
from Bookstore import app, Book

#Function to print titles in alphabetical order
def print_titles():
    with app.app_context():
        rows = Book.query.with_entities(Book.book_name).order_by(Book.book_name).all()
        for (title,) in rows:
            print(title)


if __name__ == '__main__':
    print_titles()
```
