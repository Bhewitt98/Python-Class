```python
#Imports multiprocessing to run multiple processes 
import multiprocessing
#Imports the time module so we can count the time
import time
#Imports random module so the wait time can be randomized
import random
#imports the datetime module so we can display the time each process was run
from datetime import datetime

#function to make process wait x amount of time
def random_wait():
    wait_time = random.uniform(0, 1)
    time.sleep(wait_time)
    current_time = datetime.now().strftime("%H:%M:%S.%f")
    print(f"Process {multiprocessing.current_process().name} waited for {wait_time:.3f} seconds at {current_time}")

if __name__ == "__main__":
    processes = []
    for i in range(3):
        process = multiprocessing.Process(target=random_wait)
        processes.append(process)
        process.start()
    
    for process in processes:
        process.join()
```
